```python
def f(t):
  return np.exp(-t)*np.cos(2*np.pi*t)

t1 = np.arange(0,5,0.1)
t2 = np.arange(0,5,0.02)

plt.figure()

plt.subplot(211)
plt.plot(t1,f(t1),'bo',t2,f(t2),'r')

plt.subplot(212)
plt.plot(t2,np.cos(2*np.pi*t2),'y--')

plt.show()
```

```python
mu,sigma = 100,16
x = mu+sigma*np.random.randn(10000)

n,bins,patches = plt.hist(x,50,density=1,facecolor='r',alpha = 0.5)

plt.xlabel('Smarts')
plt.ylabel('Probab')
plt.title('IQ')
plt.text(60,0.025,"mu=100,sigma=15")
plt.axis([40,160,0,0.03])
plt.grid(True)
plt.show()
```
